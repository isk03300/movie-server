```python
my_rating = pd.read_csv('../data/My_Ratings.csv')
```

```python
my_rating
```

```python
title = my_rating['Movie Name'][0]
```

```python
title
```

```python
corr_movie[title]
```

```python
recom_movie = corr_movie[title].dropna().sort_values(ascending=False).to_frame()
```

```python
recom_movie
```

```python
recom_movie.columns = ['corr']
```

```python
recom_movie['weight'] = recom_movie['corr'] * my_rating['Ratings'][0]
```

```python
recom_movie
```

```python
# 비어있는 데이터프레임 만들고, 
# 내가 본 영화의 가중치까지 계산해서, 추천영화를 뽑아내자.
# 추천영화들를 비어있는 데이터프레임에 추가하면 된다.
```

```python
movie_list = pd.DataFrame()

for i in range(my_rating.shape[0]) :
    title = my_rating['Movie Name'][i]
    recom_movie = corr_movie[title].dropna().sort_values(ascending=False).to_frame()
    recom_movie.columns = ['corr']
    recom_movie['weight'] = recom_movie['corr'] * my_rating['Ratings'][i]
    movie_list = pd.concat([movie_list,recom_movie])
    
```

```python
movie_list
```

```python
movie_list.sort_values('weight', ascending=False, inplace=True)
```

```python
# 내가 이미 본 영화는 추천에서 제외한다.

for title in my_rating['Movie Name'] :
    if title in movie_list.index :
        movie_list.drop(title, axis = 0, inplace=True)
```

```python
# 이 리스트에는, 중복된 영화도 있을 수 있다.
# 따라서 중복된 영화가 있으면, 웨이트 점수가 높은 것으로만 나오도록 한다.


movie_list.reset_index().value_counts('title')
```

```python
movie_list = movie_list.groupby('title')['weight'].max().sort_values(ascending=False).head(10)
```
